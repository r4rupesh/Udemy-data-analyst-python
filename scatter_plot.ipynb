```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data =pd.read_excel("customer_data+performance.xlsx")
data.head(0)
```

```python
numeric_data = data.select_dtypes(include=['number']).columns
```

```python
corr_data =round(data[numeric_data].corr(),3)
corr_data
```

```python
plt.figure(figsize=(6, 3))
plt.scatter(data["Frequency_of_Purchases"],data["Income"])
plt.title("Relation Between FPA and income")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.regplot(x="Frequency_of_Purchases",y="Income",data=data)
plt.title("Distribution Of Purchase Amount")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```
