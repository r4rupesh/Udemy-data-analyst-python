```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data =pd.read_excel("customer+performance+additional.xlsx")
```

```python
freq_purchase_channel=data["Purchase_Channel"].value_counts()
```

```python
freq_purchase_channel
```

```python
perc_purchase_channel=freq_purchase_channel/len(data["Purchase_Channel"])*100 
```
