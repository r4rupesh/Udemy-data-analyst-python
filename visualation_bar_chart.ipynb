```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_excel("data_3_rows.xlsx")
data.head()
```

```python
# Barchart With pyplot
```

```python
plt.figure(figsize=(8,8))
data.plot(kind="bar",x="Region",y="Average_Purchase_Amount",legend=False)
plt.title("Region, Average Purhase Amount")
plt.xlabel("Region")
plt.ylabel("Average_Purchase_Amount")
plt.xticks(rotation=45) # Rotates region names for better readability
plt.show()
```

```python
# Barchart With pyplot & seasborn
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x="Region", y="Average_Purchase_Amount", data=data)
plt.title("Region, Average Purhase Amount")
plt.xlabel("Region")
plt.ylabel("Average_Purchase_Amount")
plt.xticks(rotation=45) # Rotates region names for better readability
plt.show()
```
