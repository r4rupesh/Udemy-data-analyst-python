```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
data =pd.read_excel("customer_data.xlsx")
data.head()
```

```python
chicago_cust =data.query("City=='Chicago'")["Frequency_of_Purchases"]
newyork_cust =data.query("City=='New York'")["Frequency_of_Purchases"]
houston_cust =data.query("City=='Houston'")["Frequency_of_Purchases"]
losangeles_cust =data.query("City=='Los Angeles'")["Frequency_of_Purchases"]
```

```python
Test_stat,p_value =stats.f_oneway(chicago_cust,newyork_cust,houston_cust,losangeles_cust)
```

```python
if p_value<0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
print("P value: ",p_value)
```

```python
from scipy.stats import levene
```

```python
stats,p_value =levene(chicago_cust,newyork_cust,houston_cust,losangeles_cust)
```

```python
print("P value: ",p_value)
```
