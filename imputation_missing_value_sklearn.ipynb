```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
data= pd.read_excel("customer_data1.xlsx")
```

```python
missing_value=data.isnull().sum()
missing_value
```

```python
imputer=SimpleImputer(strategy="mean")
imputer.fit(data[["Income"]])
data[["Income"]]=imputer.transform(data[["Income"]])
```

```python
imputer=SimpleImputer(strategy="most_frequent")
imputer.fit(data[["City"]])
data[["City"]]=imputer.transform(data[["City"]])
```

```python
data.to_excel("customer_data_impute_scklearn.xlsx",index=False)
```
