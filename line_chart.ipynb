```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data =pd.read_excel("customer_data+performance.xlsx")
data.head()
```

```python
city_purchase_amount=data.groupby("City")["Average_Purchase_Amount"].median()
city_purchase_amount
```

```python
plt.figure(figsize=(6, 3))
city_purchase_amount.plot(kind="line",marker="o")
plt.title("Average Purchage Amount")
plt.xlabel("Cities")
plt.ylabel("Purchase amount")
plt.show()
```

```python
plt.figure(figsize=(6, 3))
sns.lineplot(data=city_purchase_amount,marker="o")
plt.xlabel("Cities")
plt.ylabel("Purchase frequecy")
plt.show()
```
