```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_excel("customer+performance+additional.xlsx")
```

```python
cross_tab=pd.crosstab(data["City"],data["Churn_Status"])
```

```python
cross_tab
```

```python
# stacked  bar chart
plt.figure(figsize=(10,8))
cross_tab.plot(kind="bar",stacked=True)
plt.title("Customer Who Churn By City")
plt.xlabel("City")
plt.ylabel("Frequencies")
plt.xticks(rotation=45) # Rotates region names for better readability
plt.show()
```

```python
# clustered bar chart
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x="City",hue="Churn_Status",data=data)
plt.title("Customer Who Churn By City")
plt.xlabel("City")
plt.ylabel("Frequencies")
plt.xticks(rotation=45) # Rotates region names for better readability
plt.show()
```
