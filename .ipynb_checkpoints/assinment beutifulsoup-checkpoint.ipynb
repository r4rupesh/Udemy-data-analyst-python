```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url ="https://en.wikipedia.org/wiki/Samsung"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36"
}
response = requests.get(url,headers=headers)
```

```python
html_content=response.text
```

```python
soup= BeautifulSoup(html_content,"html.parser")
```

```python
headings =[heading.text.strip() for heading in soup.find_all(["h1","h2","h3"])]
```

```python
print(headings)
```

```python
paragraphs =[p.text.strip() for p in soup.find_all(["p"])]
print(paragraphs)
```

```python
links=[a["href"] for a in soup.find_all("a",href=True)]
print(links)
```

```python
infobox = soup.find("table",class_= "infobox")
print(infobox)
```

```python
if infobox:
    rows =infobox.find_all("tr")
    data={}
    for row in rows:
        header= row.find("th")
        value =row.find("td")
        if header and value:
            data[header.text.strip()]=value.text.strip()
        print(data)
else:
    print("infobox table not found")
```

```python
table=pd.DataFrame(list(data.items()),columns=["Key","Value"])
table
```

```python
table.to_excel("samsumg_inc.xlsx")
```
