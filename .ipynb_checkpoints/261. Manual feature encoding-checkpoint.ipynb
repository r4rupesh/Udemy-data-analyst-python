```python
import pandas as pd
```

```python
data = pd.read_excel("customer_data.xlsx")
data.head()
nan_data = data[data["Customer_Lifespan_Months"].isna()]
print(nan_data)
```

```python
data["Customer_Lifespan_Months"] = pd.to_numeric(data["Customer_Lifespan_Months"], errors='coerce')
```

```python
data["engagement_level"]=pd.cut(data["Customer_Lifespan_Months"],bins=[0,2,3,5],labels=["Low engaged","Moderately engaged","Highly engaged"],include_lowest=True)
```

```python
data.head()
```

```python
mapping={"Low engaged":1,"Moderatly engaged":2,"Highly engaged":3}
data["engagement_level"]=data["engagement_level"].map(mapping)
data.head()
```

```python
data.isna().sum()
```
