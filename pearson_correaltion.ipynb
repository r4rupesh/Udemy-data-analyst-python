```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
data =pd.read_excel("customer_data.xlsx")
data.head()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(x="Frequency_of_Purchases",y="Average_Purchase_Amount",data=data)
plt.title ("linerity between FP and APA")
plt.xlabel("Frequency of Purchases")
plt.ylabel("Average Purchase Amount")
plt.show()
```

```python
purchase_frequency =data["Frequency_of_Purchases"]
purchase_amount =data["Average_Purchase_Amount"]
corr,p_value =pearsonr(purchase_frequency,purchase_amount)
print("Correlation coefficient: ",corr)
print("P value: ",p_value)

```

```python
if p_value<0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
